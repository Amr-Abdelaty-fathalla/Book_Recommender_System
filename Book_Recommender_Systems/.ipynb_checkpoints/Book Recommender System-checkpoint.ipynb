### Collaborative Filtering Using k-Nearest Neighbors (kNN)
- kNN is a machine learning algorithm to find clusters of similar users based on common book ratings, and make predictions using the average rating of top-k nearest neighbors. by presenting ratings as a in a matrix using pivot-table this matrix contains users as columns and book_title as row to see rating of all users to all books and this will simplify calcuations of similarity between users   
- We convert our table to a 2D matrix, and fill the missing values with zeros (since we will calculate distances between rating vectors). We then transform the values(ratings) of the matrix dataframe into a scipy sparse matrix for more efficient calculations.

- Finding the Nearest Neighbors We use unsupervised algorithms with sklearn.neighbors. The algorithm we use to compute the nearest neighbors is brute, and we specify metric=cosine so that the algorithm will calculate the cosine similarity between rating vectors

```python
# Create pivot table
book_pivot = final_ratings.pivot_table(columns = 'User-ID' , index = 'Title' , values = 'Rating')
```

```python
book_pivot
```

```python
# NaN here means this user give rate to this book '0' --> convert Nan to 0
book_pivot.fillna(0, inplace = True)
```

```python
book_pivot
```

```python
# we don't care about zeros when we calculate distance with respect to all columns and all rows
# we will CSR system : automatically won't consider 0 in calculations will do to find similarities between users
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(book_pivot)
```

```python
book_sparse 
```

## KNN Model

```python
from sklearn.neighbors import NearestNeighbors
```

```python
knn_model = NearestNeighbors(algorithm='brute' , metric = 'cosine')
```

```python
knn_model.fit(book_sparse)
```

```python
# select one book and try to find out the nearest 'k' neighbors from this book using knn_model we trained on our data
book_query_index = np.random.choice(book_pivot.shape[0])

print(book_query_index) # Book_index
print(book_pivot.index[book_query_index]) # Book_name
print(book_pivot.iloc[book_query_index , :].values) # Book_values

distance , indices = knn_model.kneighbors(book_pivot.iloc[book_query_index , :].values.reshape(1,-1), n_neighbors = 6) 
```

```python
# to see name of these books
book_names = []
for i in range(len(indices)):
    book_names.append(book_pivot.index[indices[i]])
```

```python
book_names
```

```python
def recommend(book_name):
    # return index of book in pivot table
    book_id = np.where(book_pivot.index == book_name)[0][0]
    distance , indices = knn_model.kneighbors(book_pivot.iloc[book_id , :].values.reshape(1,-1), n_neighbors = 6) 

    for i in range(distance.flatten().shape[0]):
        if i == 0:
            print(f'recommendation for this book "{book_pivot.index[book_id]}" :\n')
        else:
            print(f'{i} ==> {book_pivot.index[indices.flatten()[i]]} with distance : {distance.flatten()[i]}')
    
```

```python
book_name = 'Harry Potter and the Chamber of Secrets (Book 2)'
recommend(book_name)
```
